```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew
from math import sqrt
```

```python
def draw_until_6():
    count = 0
    val = None
    while val != 6:
        val = np.random.randint(1, 7)
        count += 1

    return count

def mean(x):
    return sum(x) / len(x)

def variance(x):
    return sum([(xi - mean(x))**2 for xi in x]) / (len(x) - 1) 

def std(x):
    return sqrt(variance(x))

def mean_absolute_deviation(x):
    avg = mean(x)
    return sum([np.abs(xi - avg) for xi in x]) / len(x)

def skewness(x):
    avg = mean(x)
    x_std = std(x)
    return (sum([(xi - avg)**3 for xi in x])) / ((len(x) - 1) * x_std**3)

def sum_n_draws(n):
    return sum([np.random.randint(1, 7) for _ in range(n)])
```

```python
tries = 10_000
counts = [draw_until_6() for _ in range(tries)]

counts_mean = mean(counts)
counts_std = std(counts)
counts_var = variance(counts)
counts_skew = skew(counts)
counts_mean_absolute_deviation = mean_absolute_deviation(counts)
```

```python
plt.figure(figsize=(8, 4))
y, _, _ = plt.hist(counts, bins=50, ec='blue', fc='navy', alpha=0.5)
plt.vlines(counts_mean, ymin=0, ymax=y.max(), colors='r', ls='--', label=f'mean=${round(counts_mean, 3)}$')
plt.xlabel('Number of tries')
plt.ylabel('Frequency')
plt.title(f'Histogram of number of tries\nstd=${round(counts_std, 3)}$, variance=${round(counts_var, 3)}$, skewness=${round(counts_skew, 3)}$,\nmean absolute deviation (m(X) = mean)=${round(counts_mean_absolute_deviation, 3)}$')
plt.legend(loc='upper right')
plt.show()
```
