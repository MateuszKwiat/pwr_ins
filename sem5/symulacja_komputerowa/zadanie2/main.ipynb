```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_until_6():
    count = 0
    val = None
    while val != 6:
        val = np.random.randint(1, 7)
        count += 1

    return count
```

```python
tries = 10_000
counts = [draw_until_6() for _ in range(tries)]

counts_mean = np.mean(counts)
counts_std = np.std(counts)
counts_var = np.var(counts)
```

```python
plt.figure(figsize=(8, 4))
y, _, _ = plt.hist(counts, bins=50, ec='blue', fc='navy', alpha=0.5)
plt.vlines(counts_mean, ymin=0, ymax=y.max(), colors='r', ls='--', label=f'mean=${round(counts_mean, 3)}$')
plt.xlabel('Number of tries')
plt.ylabel('Frequency')
plt.title(f'Histogram of number of tries\nstd=${round(counts_std, 3)}$, variance=${round(counts_var, 3)}$')
plt.legend(loc='upper right')
plt.show()
```
